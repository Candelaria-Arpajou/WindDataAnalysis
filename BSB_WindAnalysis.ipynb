```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.express as px
import scipy.stats as s
from datetime import date, datetime
from google.colab import files
```

```python
uploaded = files.upload()
file_name_BSB = 'Bahia_San_Blas_2019-03-21_2022-03-20.csv'
Bahia_San_Blas_df = pd.read_csv(file_name_BSB, skiprows=range(24), header = 0)
```

```python
BSB_df_copy = Bahia_San_Blas_df.copy(deep = True)
BSB_df_copy.head(5)
```

```python
sns.jointplot(data=BSB_df_copy, x='T2M',y='WS50M', kind='hex', height=6) 
plt.axvline(BSB_df_copy['T2M'].mean(), color='orangered',linestyle='--', label='Temp media')
plt.axhline(BSB_df_copy['WS50M'].mean(), color='blue',linestyle='--', label='Velocidad media')
sns.jointplot(data=BSB_df_copy, x='PS',y='WS50M', kind='hex', height=6) 
plt.axvline(BSB_df_copy['PS'].mean(), color='orangered',linestyle='--', label='Presion media')
plt.axhline(BSB_df_copy['WS50M'].mean(), color='blue',linestyle='--', label='Velocidad media')
plt.legend()
# Remove scietific notation
plt.ticklabel_format(style='plain', axis='y')
```
